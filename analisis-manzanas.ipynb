## Tratamiento de Nan

```python
np.mean(moscu)
```

```python
np.mean(kaliningrado)
```

```python
plt.plot(dates, kaliningrado)
```

```python
np.sum(np.isnan(kaliningrado))
```

Solucion propuesta: a pesar de que para este caso en particular no resulta necesario, me parecio pertinente hacerlo tambien de una forma en la que el codigo pueda ser escalable a interpolaciones para datasets con mayor numero de datos NaN

```python
kaliningrado_nans = np.isnan(kaliningrado)
kaliningrado_not_nans = ~ kaliningrado_nans

kaliningrado[kaliningrado_nans] = np.interp(np.flatnonzero(kaliningrado_nans), np.flatnonzero(kaliningrado_not_nans), kaliningrado[kaliningrado_not_nans])
```

```python
np.mean(kaliningrado)
```

## Diferencias entre arrays

```python
plt.plot(dates, moscu)
```

```python
x = dates
y = 0.5 * x + 80
plt.plot(dates, moscu)
plt.plot(x, y)
```

```python
np.sqrt(np.sum(np.power(moscu - y, 2)))
```

```python
np.linalg.norm(moscu - y)
```

En lugar de simplemente proponer una ecuaci칩n lineal, quisiera realizar el calculo de una utilizando las herramientas de algebra lineal de numpy

```python
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, moscu, rcond=None)[0]
new_y = m * x + c
plt.plot(dates, moscu)
plt.plot(dates, new_y)
np.linalg.norm(moscu - new_y)
```

Curiosamente, el resultado fue muy muy similar, pero el uso de esto no es simplemente tener el resultado, si no construir un m칠todo eficaz.
En esta segunda solucion, se utilizo el enfoque mediante polyfit

```python
m, b = np.polyfit(dates, moscu, 1)
y = m * dates + b
plt.plot(dates, moscu)
plt.plot(dates, y)
np.linalg.norm(moscu - y)
plt.plot(24, m*25 + b, '*r')
```

```python
m_orange, b_orange = np.polyfit(data_orange_diameter, data_orange_weight, 1)
m_grapefruit, b_grapefruit = np.polyfit(data_grapefruit_diameter, data_grapefruit_weight, 1)

prediction_orange = m_orange * data_orange_diameter + b_orange
prediction_grapefruit = m_grapefruit * data_grapefruit_diameter + b_grapefruit

plt.plot(data_orange_diameter, data_orange_weight, 'r')
plt.plot(data_grapefruit_diameter, data_grapefruit_weight, 'b')
plt.plot(data_orange_diameter, prediction_orange, 'm')
plt.plot(data_grapefruit_diameter, prediction_grapefruit, 'c')

plt.legend(['orange', 'grapefruit', 'prediction_orange', 'prediction_grapefruit'])
```

### Ejercicio
**Calculo de coeficientes angulares para el caso de las naranjas y toronjas.**

Se usaran numeros aleatorios para calcular distintas pendientes para la funcion lineal de naranja y toronja.


```python
np.random.seed(58)
rand_slope_values = np.random.uniform(10, 20, 100)
print(rand_slope_values)
```

```python
norm_orange = np.array([])
norm_grapefruit = np.array([])
constant_orange_grapefruit = 17

for i in range(len(rand_slope_values)):
    norm_orange = np.append(norm_orange, np.linalg.norm(data_orange_weight - (data_orange_diameter * rand_slope_values[i] + constant_orange_grapefruit)))
    norm_grapefruit = np.append(norm_grapefruit, np.linalg.norm(data_grapefruit_weight - (data_grapefruit_diameter * rand_slope_values[i] + constant_orange_grapefruit)))

print(norm_orange)
print(norm_grapefruit)

```

```python
rand_norm_orange_grapefruit = np.column_stack((rand_slope_values, norm_orange, norm_grapefruit))
print(rand_norm_orange_grapefruit)
```

```python
np.savetxt('rand_norm_orange_grapefruit.csv', rand_norm_orange_grapefruit, delimiter=',')
```
